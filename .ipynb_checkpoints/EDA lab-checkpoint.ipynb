__Part 3__: Unpacking Product Ratings
Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

Measures of Centrality:

__4) Calculate the mean, median, and mode for the rating of products.
How do customers generally rate products? Is there a common trend?__




```python
df_rating_mean = df["stars"].mean()
df_rating_mod = df["stars"].mode()[0]
df_rating_median = df["stars"].median()
print(f"Rating mean value = {df_rating_mean}")
print(f"Rating mode = {df_rating_mod}")
print(f"Rating median value = {df_rating_median} ")

Q3_stars = df['stars'].quantile(0,75)
Q3_stars
```

Measures of Dispersion:

__4) Determine the variance, standard deviation, and interquartile range for product rating.
Are the ratings consistent, or is there a wide variation in customer feedback?
Shape of the Distribution:__

```python
stars_std = df["stars"].std()
stars_var = df["stars"].var()
quantiles_stars = df['stars'].quantile([0.25, 0.5, 0.75])
```

```python
print( stars_std)
print (stars_var)
print (quantiles_stars)
```

__5) Calculate the skewness and kurtosis for the rating column.
Are the ratings normally distributed, or do they lean towards higher or lower values?
Visualizations:
Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?__

```python
skewness_stars = df['stars'].skew()
kurtosis_stars = df['stars'].kurtosis()

skewness_stars, kurtosis_stars
```

```python
 # Positive skewness_stars: the distribution is right-skewed.
#Distributions with negative kurtosis have light tails and a flatter peak. This means they have fewer outliers than a normal distribution.
```
